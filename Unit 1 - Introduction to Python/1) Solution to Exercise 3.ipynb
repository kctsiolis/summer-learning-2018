To show how quickly the series converges, I took input for partial sums calculated with 2, 5, and 10 terms. With that, you can see the plots of how the approximation improves and how it compares to the actual graph of cos(x).

```
import pylab as pl
import numpy as np 

k = int(input("Enter the number of terms k: "))

def approx(x,k):
  sum = 0
  for n in range(k): 
    sum += ((-1)**n*(x**(2*n)))/(np.math.factorial(2*n))
  return sum

cos = lambda x: approx(x,k)
x = np.arange(0,2*np.pi,0.01) 
y = cos(x)
pl.ion() 
pl.plot(x,y)
pl.xlabel('x')
pl.ylabel('y')
pl.title("Cosine Function approximated by Taylor Series with " + str(k) + " terms")
pl.show()
```
