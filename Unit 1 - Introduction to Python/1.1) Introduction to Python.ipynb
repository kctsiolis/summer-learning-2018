From infinite series in Cal II, remember that computing large partial sums allows for good approximations of functions, among other things. Before we get to approximating functions, here is a simple example of computing partial sums.

```
n = int(input("n: "))
sum = 0

for i in range(n+1):
  sum += i #At each step, we add a new term to the partial sum

print("The " + str(n) + "th partial sum is: " + str(sum))
```

How does a computer determine the values of exponential and trigonometric functions with high precision? Using Taylor series!

```
import numpy as np

sum = 0
x = 1
actual = np.exp(1) #This is e!

for n in range(20): #20 steps
  sum += ((x)**n)/(np.math.factorial(n)) #Power series for e^x
  
error = np.abs(sum - actual) #np.abs() will return the absolute value of the argument
percent = error/actual

print("Taylor approximation: " + str(sum))
print("Actual value: " + str(actual))
print("Error: " + str(error))
print("Percent Error: " + str(percent))
```

```
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))

def quadFormula(a,b,c): #A quadratic formula function that takes a, b, and c as inputs
  disc = b**2 - 4*a*c #The discriminant
  if disc > 0: #Case of 2 real roots
    x1 = (-b - disc**0.5)/2*a
    x2 = (-b + disc**0.5)/2*a
    print("The solutions are: " + str(x1) + " and " + str(x2))
  elif disc == 0: #Case of 1 real root
    x = (-b + disc**0.5)/2*a
    print("The solution is: " + str(x))
  else: #Case of no real roots
    print("No real solution.")
    
quadFormula(a,b,c)
```

Using Matplotlib to plot functions!

```
import pylab as pl
import numpy as np #If you have already imported numpy in a previous cell, this is redundant

sin = lambda x: np.sin(x) #The "lambda" makes it so that we compute the sin at each value of x
x = np.arange(0,2*np.pi,0.01) #Sets scale for x-axis (start at x = 0, end at x = 6.28, compute the value of y in steps of 0.01)
y = sin(x)
pl.ion() #Turns on interactive plotting
pl.plot(x,y) #Plots x and y (in this case, y as a function of x)
pl.xlabel('x')
pl.ylabel('y')
pl.title('Sine Function')
pl.show()
```

```
import pylab as pl
import numpy as np 

def approx(x):
  sum = 0
  for n in range(10): 
    sum += ((x)**n)/(np.math.factorial(n))
  return sum

e = lambda x: approx(x)
x = np.arange(-2,2,0.01) 
y = e(x)
pl.ion() 
pl.plot(x,y)
pl.xlabel('x')
pl.ylabel('y')
pl.title('Exponential Function approximated by Taylor Series with 10 terms')
pl.show()
```

Python is an object-oriented programming language. This means that every variable contains an object, and each object is an instance of a class. A class comes with a set of attributes and methods that can be performed on objects of that class.

```
class myClass():
  def __init__(self):
    self.a = 3
    self.b = 2
    
  def function(self):
    x = self.a + self.b
    return x

g = myClass().function()
print(g)

h = myClass().function()
print(g * h)
```

You can also create new classes that inherit from previously established classes. These can be viewed as "children" of "parent" classes.

```
class myNewClass(myClass):
  def __init__(self):
    super().__init__() #Initializes superclass, the class that we inherit from
    #Because of the superclass, we also have a = 3 and b = 2
    self.c = 4
  def newFunction(self):
    y = myClass().a + myClass().b + self.c #3 + 2 + 4
    return y

j = myNewClass().newFunction()
print(j)
```
