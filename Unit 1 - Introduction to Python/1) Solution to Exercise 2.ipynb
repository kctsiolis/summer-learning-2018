```
prices = [29.95, 74.45, 19.75, 23.55, 225.00] #Making a list of prices is very useful

print("Welcome to my very exclusive sports store! We have 5 items for sale: ")
print("1. World Cup Soccer Ball Replica - $29.95")
print("2. Soccer Cleats - $74.45")
print("3. Shin Pads - $19.75")
print("4. Sports Bag - $23.55")
print("5. Signed Cristiano Ronaldo Framed Picture - $225.00")
```

We have to be clever and efficient about how we take user input because the user may want to buy more than one of the five items (and more than one of each).

```
answer = "Y"
total = 0

while answer == "Y":
  item = int(input("Which one of these items would you like to purchase? "))
  while item < 1 or item > 5:
    item = int(input("Sorry. That input was invalid. Please try again. "))
  qty = int(input("How many of this item do you want? "))
  while qty <= 0:
    qty = int(input("Sorry. That input was invalid. Please try again. "))
  total += prices[item - 1] * qty
  
  answer = input("Would you like to purchase more items? (Y/N) ")
  while answer != "Y" and answer != "N":
    answer = input("Sorry. That input was invalid. Please answer Y or N. ")
```

```
print("Your total is $" + str("%.2f" % total) + ".") #"%.2f" is used to ensure that printing total always shows two decimal places

cash = float(input("Cash Payment: $"))
while cash < total:
  cash = float(input("Please ensure that you provide an amount greater or equal to $" + str(total) + ": $"))
```

The following is my method of calculating the change. There are other ways to do it, including a way that uses modular division (which I mentioned as a hint when I presented the exercise). However, I later realized this is a better and probably more intuitive way of doing it. Sorry for throwing anyone off.

```
if cash != total:
  change = cash - total
  print("You will be given $" + str("%.2f" % change) + " in change.")
  
  hundreds = change // 100
  change -= hundreds*100
  change = round(change,2) #Rounds change to two decimal places (prevents errors with integer division)
  
  fifties = change // 50
  change -= fifties*50
  change = round(change,2)
  
  twenties = change // 20
  change -= twenties*20
  change = round(change,2)
  
  tens = change // 10
  change -= tens*10
  change = round(change,2)
  
  fives = change // 5
  change -= fives*5
  change = round(change,2)
  
  toonies = change // 2
  change -= toonies*2
  change = round(change,2)
  
  loonies = change // 1
  change -= loonies
  change = round(change,2)
  
  quarters = change // 0.25
  change -= quarters*0.25
  change = round(change,2)
  
  dimes = change // 0.10
  change -= dimes*0.10
  change = round(change,2)
  
  nickels = change // 0.05
  
  print("Hundred Dollar Bills: " + str(int(hundreds)))
  print("Fifty Dollar Bills: " + str(int(fifties)))
  print("Twenty Dollar Bills: " + str(int(twenties)))
  print("Ten Dollar Bills: " + str(int(tens)))
  print("Five Dollar Bills: " + str(int(fives)))
  print("Toonies: " + str(int(toonies)))
  print("Loonies: " + str(int(loonies)))
  print("Quarters: " + str(int(quarters)))
  print("Dimes: " + str(int(dimes)))
  print("Nickels: " + str(int(nickels)))
```
