The strategy that ensures you always win this game is to keep guessing the number in the centre of the interval of all possible numbers. This will partition the possibilities in a very efficient way.

```
import random

secretNumber = random.randint(0,100) #Generates a random integer

print("Welcome to the Guessing Game! I'm thinking of a number between 0 and 100! You've got seven tries to guess it!")

for i in range(7):
  guess = int(input())
  if guess == secretNumber:
      print("Congratulations! You have guessed the secret number in " + str(i+1) + " guesses!")
      break #This statement causes the program to exit the for loop before 7 iterations are reached
  elif guess < 0 or guess > 100:
      print("That was not a valid guess. Please try again. And just because I'm mean, I'm " +
            "still counting that as a guess. You have " + str(6 - i) + " guesses remaining.")
  else:
      if guess < secretNumber:
          if i < 5:
              print("Guess higher. You have " + str(6 - i) + " guesses remaining.")
          if i == 5:
              print("Guess higher. You have 1 guess remaining.")
      else:
          if i < 5:
              print("Guess lower. You have " + str(6 - i) + " guesses remaining.")
          if i == 5:
              print("Guess lower. You have 1 guess remaining.")
  if i == 6:
      print("That makes ten guesses. You lose. The secret number was " + str(secretNumber) + ".")
```

This is a solution to Optional Exercise 1. Using the try/except block, this ensures that no invalid input gets through and causes an error.

```
import random

secretNumber = random.randint(0,100)

print("Welcome to the Guessing Game! I'm thinking of a number between 0 and 100! You've got seven tries to guess it!")

for i in range(7):
    try: #If you don't know what "try" means, it's ok (this is just a bonus exercise for those who are familiar)
        guess = int(input())
        if guess == secretNumber:
            print("Congratulations! You have guessed the secret number in " + str(i+1) + " guesses!")
            break
        elif guess < 0 or guess > 100:
            print("That was not a valid guess. Please try again. And just because I'm mean, I'm still counting that as a guess. You have " + str(6 - i) + " guesses remaining.")
        else:
            if guess < secretNumber:
                if i < 5:
                    print("Guess higher. You have " + str(6 - i) + " guesses remaining.")
                if i == 5:
                    print("Guess higher. You have 1 guess remaining.")
            else:
                if i < 5:
                    print("Guess lower. You have " + str(6 - i) + " guesses remaining.")
                if i == 5:
                    print("Guess lower. You have 1 guess remaining.")
        if i == 6:
            print("That makes ten guesses. You lose. The secret number was " + str(secretNumber) + ".")
    #Bonus
    except ValueError: 
        print("Error. Your entry was invalid.")
        if i < 5:
                print("You have " + str(6 - i) + " guesses remaining.")
        if i == 5:
                print("You guessed wrong. Try again. You have 1 guess remaining.")
        if i == 5:
            print("You guessed wrong. That makes ten guesses. You lose. The secret number was " + str(secretNumber) + ".")
```
